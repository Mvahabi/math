# Calculator
Best version I made so far using classes and objects

```python
class Calc:
    def add(self, *numbers):
        return sum(numbers)

    def sub(self, *numbers):
        result = numbers[0]
        for num in numbers[1:]:
            result -= num
        return result

    def mult(self, *numbers):
        result = 1
        for num in numbers:
            result *= num
        return result

    def div(self, *numbers):
        if 0 in numbers[1:]:
            return "Cannot divide by zero"
        result = numbers[0]
        for num in numbers[1:]:
            result /= num
        return result

    def cmp(self, x, y):
        if x < y:
            return f"{x} is smaller than {y}\n"
        elif x > y:
            return f"{x} is greater than {y}\n"
        else:
            return "Both numbers are equal\n"

    def mod(self, x, y):
        return x % y

    def power(self, x, y):
        return x ** y

    def mean(self, *numbers):
        if len(numbers) == 0:
            return None
        return sum(numbers) / len(numbers)

print("Welcome to the calculator app!\n")
opr = Calc()
choice = 1
while choice != 9:
    print("1. addition")
    print("2. subtraction")
    print("3. multiply")
    print("4. divide")
    print("5. modulus")
    print("6. power")
    print("7. min/max")
    print("8. mean")
    print("9. exit")

    choice = input("Enter choice: ")
    print()
    if choice == 9:
        print("good bye!")
        break
    try:
        choice = int(choice)
        if (choice < 1) or (choice > 9):
            print("Error!!!")
            print()
        elif choice == 1:
            print("Operation: Add")
            num_list = input("Enter the numbers separated by spaces: ").split()
            numbers = [float(num) for num in num_list]
            print(opr.add(*numbers))
            print()

        elif choice == 2:
            print("Operation: Sub")
            num_list = input("Enter the numbers separated by spaces: ").split()
            numbers = [float(num) for num in num_list]
            print(opr.sub(*numbers))
            print()

        elif choice == 3:
            print("Operation: Mult")
            num_list = input("Enter the numbers separated by spaces: ").split()
            numbers = [float(num) for num in num_list]
            print(opr.mul(*numbers))
            print()

        elif choice == 4:
            print("Operation: Div")
            num_list = input("Enter the numbers separated by spaces: ").split()
            numbers = [float(num) for num in num_list]
            print(opr.div(*numbers))
            print()

        elif choice == 5:
            print("Operation: Mod")
            x = int(input("Enter num1: "))
            y = int(input("Enter num2: "))
            print(opr.mod(x, y))
            print()

        elif choice == 6:
            print("Operation: Power")
            x = int(input("Enter num1: "))
            y = int(input("Enter num2: "))
            print(opr.power(x, y))
            print()

        elif choice == 7:
            print("Operation: min/max")
            x = int(input("Enter num1: "))
            y = int(input("Enter num2: "))
            print(opr.cmp(x, y))
            print()
        elif choice == 8:
            print("Operation: AVG")
            num_list = input("Enter the numbers separated by spaces: ").split()
            numbers = [float(num) for num in num_list]
            print(opr.mean(*numbers))
            print()
    except ValueError:
        print("Can not take sting as a choice, Try again!\n")


# c = Calc()
# assert c.add(1, 2, 3) == 6
# assert c.sub(1, 2, 3, -4) == 0
# assert c.mult(1, 2, 3, -4) == -24
# assert c.div(100, 2, 5, 2) == 5.0
# assert c.cmp(3, 5) == "3 is smaller than 5\n"
# assert c.cmp(5, 3) == "5 is greater than 3\n"
# assert c.cmp(5, 5) == "Both numbers are equal\n"
# assert c.mod(8, 91) == 8
# assert c.mod(91, 8) == 3
# assert c.power(3,3) == 27
# assert c.mean(1,2,3,4,5) == 3.0
# print("Passed all tests!")
```
